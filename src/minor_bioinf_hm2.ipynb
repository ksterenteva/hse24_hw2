* Индексируем белки T.oleivorans MIL-1:

На выходе получаем файлы T_oleivorans_MIL_1.proteins.phr, T_oleivorans_MIL_1.proteins.pin, T_oleivorans_MIL_1.proteins.psq

## Определяем функции белков нашей бактерии через поиск по БД SwissProt


* Смотрим, какие белки нашей бактерии (из Баренцева моря) не имели похожих белков из бактерии T.oleivorans MIL-1* Для этого получаем список уникальных номеров генов бактерии из Барнцева моря, коротые имели хотя бы один похожий белок бактерии из Сицилли

Извлекаем уникальные идентификаторы белков, у которых есть совпадения с белками из бактерии T. oleivorans MIL-1 и сохраняем их в отдельный файл:

```python
!cut -f 1 scaffolds.hits_from_MIL_1.txt | sort -n | uniq > proteins.with_hits_from_MIL_1.txt
```

Используем seqkit для поиска белков, у которых нет совпадений с белками из бактерии T. oleivorans MIL-1 и сохраняем их в отдельный файл:

```python
!./seqkit grep  --invert-match  -f proteins.with_hits_from_MIL_1.txt  proteins.fasta   -o proteins.without_MIL_1.fasta
```

* Проверяем, что кол-ва белков сходятся
* Общее кол-во белков в нашей бактерии:

```python
!grep '>' proteins.fasta | wc -l
```

* Кол-во белков, которые схожи с белками из бактерии MIL-1

```python
!wc -l proteins.with_hits_from_MIL_1.txt
```

```python
!grep '>' proteins.without_MIL_1.fasta | wc -l
```

* Сравнить все белки (протеом) нашей бактерии с аннотированными белками из БД SwissProt
* Индексируем белки SwissProt:
